```python
data = {
    'product_id': amazon_data['product_id'],
    'discount_percentage':amazon_data['discount_percentage']  # Including an empty string to illustrate the issue
}

rf = pd.DataFrame(data)

# Convert to string
rf['discount_percentage'] = rf['discount_percentage'].astype(str)

# Remove non-numeric characters
rf['discount_percentage'] = rf['discount_percentage'].str.replace('[^\d.]', '', regex=True)

# Replace empty strings with '0'
rf['discount_percentage'] = rf['discount_percentage'].replace('', '0')

# Convert to float and divide by 100 to get percentage
rf['discount_percentage'] = rf['discount_percentage'].astype(float) / 100.0
df['discount_percentage'] =rf['discount_percentage']*100
# Display the cleaned data
print(rf)
```
